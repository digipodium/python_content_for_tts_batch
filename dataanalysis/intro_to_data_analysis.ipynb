# line plot 
- use to show ups and downs in s sequence or time series

```python
irisdf['sepal_length'].plot(figsize=(15,5))
```

```python
irisdf['sepal_length'].plot(figsize=(15,5),style='g--')
plt.show()
```

# scatter plot
- to check the correlation of 2 or 3 series in a dataset

```python
irisdf.plot(kind='scatter',x ='petal_length',y='petal_width')
plt.show()
```

```python
irisdf.plot(kind='scatter',x ='petal_length',y='petal_width',
            c='petal_length',cmap='cool',marker='^',alpha=.5,
            s=100,title='compairing petals dimensions' , figsize=(10,10))
plt.xticks([])
plt.show()
```

# pie chart
- used for analysing data based on percentage,
- only takes a single series of data

```python
automobile_df['engine-location'].unique()
```

```python
automobile_df['body-style'].unique()
```

```python
automobile_df['engine-location'].value_counts().plot(kind='pie')
```

```python
automobile_df['body-style'].value_counts()
```

```python
automobile_df['body-style'].value_counts().plot(kind='pie',autopct='%.2f%%')
```

```python
automobile_df['body-style'].value_counts().plot(
        kind='pie',
        autopct='%.2f%%',
        explode=[0,0,0,0.1,0.1],
        colors = ['#ff9900','#f90909','#f85323','#ffff00','#ff8833','#ff3384'])
```

```python
# donut from pie
automobile_df['body-style'].value_counts().plot(
        kind='pie',
        autopct='%.2f%%',
        wedgeprops={'width':.5},
        pctdistance=.75,
        shadow=True)
plt.show()
```

```python
automobile_df.corr().plot(kind='kde')
```
