```python
outlined_img = img.filter(ImageFilter.CONTOUR)
outlined_img.save('outlined_django.jpg')
```

```python
img.filter(ImageFilter.FIND_EDGES)
```

```python
img.filter(ImageFilter.EDGE_ENHANCE)
```

```python
img_edit = ImageEnhance.Color(img)
en_image = img_edit.enhance(-5)
```

```python
for i in range(-5,6):
    img = img.resize(size=(img.width*2,img.height*2),resample=Image.LINEAR)
    img_edit = ImageEnhance.Color(img)

    en_image = img_edit.enhance(i)
    en_image.show()
```
